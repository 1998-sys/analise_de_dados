```python
import matplotlib.pyplot as plt
```

```python
# Dados ficitícios - vendas ao longo dos meses
meses = ['Jan','Fev', 'March', 'Abril', 'Mai', 'Jun']
valores = [150,200,180,300,250,400]
```

```python
# Criando o gráfico de linha
plt.figure(figsize=(8,5))
plt.plot(
    meses,
    valores,
    marker = 'o',
    color = 'blue',
    label = 'Vendas'
)



plt.show()
```

```python
# Adicionando rótulos e títulos ao gráfico
plt.figure(figsize=(8,5))
plt.plot(
    meses,
    valores,
    marker = 'o',
    color = 'blue',
    label = 'Vendas'
)

plt.xlabel('Mês'),
plt.ylabel('Vendas'),
plt.title('Vendas ao longo dos Meses')
plt.legend()
plt.grid(True)


plt.show()
```

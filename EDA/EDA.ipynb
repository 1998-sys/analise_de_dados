# Teste de Qui-Quadrado

- h0 (hipótese Nula): As duas variáveis são independentes.
- h1 (Hipóstese alternativa): As duas variáveis não são independentes

- No teste, o objetivo é confirmar ou recusar a hipótese nula.
- Quando o valor de probabilidade H0 for menor que 0.05 (p-value) recusamos a hipótese nula e aceitamos a complementar

```python
df_tabcont  = pd.crosstab(df_churn['Churn'], df_churn['Contract'])
df_tabcont
```

```python
from scipy.stats import chi2_contingency
```

```python
chi_scores = chi2_contingency(df_tabcont)
```

```python
chi_scores
```

```python
# Número decimal de uma notação científica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
scores_churn = pd.Series(chi_scores[0])
pvalues_churn = pd.Series(chi_scores[1])
```

```python
df_chi_churn = pd.DataFrame({
    'Qui2': scores_churn,
    'Pvalue': pvalues_churn
})
```

```python
df_chi_churn
```

- As variáveis não são independentes. E pelo Chi2 alto, podemos afirmar que há uma alta correlação entre as variáveis.

### Técnica de IQR

- 1° quartil e 3° Quartil | 3° quartil - 1° quartil

```python
q1_total = df_churn_month_to['TotalCharges'].quantile(0.25)
q3_total = df_churn_month_to['TotalCharges'].quantile(0.75)
```

```python
print(q1_total)
```

```python
iqr = q3_total - q1_total
print(iqr)
```

```python
lim_inf = q1_total - (iqr * 1.5)
lim_sup = q3_total + (iqr * 1.5)
```

```python
lim_inf, lim_sup
```

```python
# Há 200 registros de outliers
len(df_churn_month_to[(df_churn_month_to.TotalCharges < lim_inf) | (df_churn_month_to.TotalCharges > lim_sup)])
```
